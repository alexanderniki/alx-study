```python
# Читаем файлы

coderun: DataFrame = pd.read_csv(
    "data/data/itresume-coderun.csv",
    delimiter=",",
    header=0
)

users: DataFrame = pd.read_csv(
    "data/data/itresume-users-pandas.csv",
    delimiter=";",
    header=1
)
```

```python
# Смотрим, что там есть
coderun.head()
```

```python
# На всякий случай, посмотрим, порядок следования айдишников пользователей
coderun.sort_values(by="user_id", ascending=True)
```

```python
# Смотрим, что там есть
users.head()
```

```python
# На всякий случай, посмотрим, порядок следования айдишников пользователей
users.sort_values(by="id", ascending=True)
```

```python
# Мержим аналогично LEFT JOIN

df: DataFrame = coderun.merge(
    users, 
    left_on="user_id", 
    right_on="id", 
    how="left"
)
df.sort_values(by="user_id")
```

```python
df.rename(columns={
    "id_x": "id_run",
    "id_y": "id_user"
})
```

```python
# сколько дней прошло между последним выполнением кода конкретного юзера и днем его регистрации. Столбец с разницей должен называться diff.

processed: DataFrame = df.groupby(["user_id"]).count()
processed
```
