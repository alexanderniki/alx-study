```python
df: pd.DataFrame = pd.read_csv(
    "data/data/itresume-users-pandas.csv", 
    delimiter=";",
    header=1
)

registrations_by_weekday: Dict = {}

df2 = df.copy()

df2['date_joined'] = pd.to_datetime(df2['date_joined'])
df2['date_joined'] = df2['date_joined'].dt.day_name()

registrations_by_weekday: Dict = (
    df2.groupby('date_joined')['id']
        .count()
        .sort_values(ascending=False)
        .to_dict()
)

print(registrations_by_weekday)
print(df)
```
