**Манипуляции со столбцами**

```python
# Проверим количество строк и столбцов
print("Строки, поля:", df.shape)

# Либо же можем посмотреть количество 
# элементов = количество строк * количество столбцов
print("Количество элементов:", df.size)

# Посмотрим на индексы в таблицы
print("Информация об индексах:", df.index)

# Посмотрим на поля, которые есть в таблице
print("Поля в таблице:", df.columns)
```

```python
# Оставим только нужные столбцы
# Порядок следования столбцов можно менять

cols: List[str] = [
    'DR_Dat',
    'DR_Tim',
    'DR_NChk',
    'DR_NDoc',
    'DR_Apt',
    'DR_NDrugs',
    'DR_Kol',
    'DR_CZak',
    'DR_CRoz',
    'DR_SDisc',
    'DR_TPay',
    'DR_CDrugs',
    'DR_Suppl',
    'DR_CDisc',
    'DR_BCDisc',
    'DR_TabEmpl',
    'DR_VZak',
    'DR_Pos'
]

df = df[cols]
print(df.shape)
df.head()
```

```python
# Переназовем столбцы более удобным образом

df.columns = [
    'dt',
    'c_time',
    'nchk',
    'ndoc',
    'apt',
    'ndrugs',
    'kol',
    'zak',
    'roz',
    'disc',
    'pay_type',
    'drug_id',
    'suppl',
    'disc_id',
    'disc_barcode',
    'empl',
    'vzak',
    'pos'
]

df
```

**Работа с типами данных**

```python
# Посмотрим типы данных
df.dtypes
```

```python
# Попробуем поменять тип disc_barcode

# Сначала проверим еще раз его тип (просто потому что можно)
print("Тип disc_barcode:", df['disc_barcode'].dtype)

# Теперь поменяем его на str
df['disc_barcode'] = df['disc_barcode'].astype('str').replace('\\.0', '', regex=True)
df.head()
```

```python
df.dtypes
```

```python
# Теперь приведем disc_id к int
df['disc_id'] = df['disc_id'].astype('str').replace('\\.0', '', regex=True)
df.head()
```

```python
# Поработаем с датой
df['dt'] = pd.to_datetime(df['dt']).dt.strftime('%d.%m.%Y')
df.head()
```

```python
# Заменим целочисленные значения кода операции на текст
df['vzak'].unique()
df['vzak'] = df['vzak'].astype('str').replace('1', 'Обычный').replace('2', 'Интернет')
df.head()
```
