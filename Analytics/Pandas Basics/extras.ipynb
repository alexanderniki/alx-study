## Merge датавфреймов

Это аналог join-а из SQL

```python
abc: DataFrame = pd.DataFrame({
    "name": ["кабачок", "помидор", "огурец"],
    "abc": ["A", "A", "C"]
})

xyz: DataFrame = pd.DataFrame({
    "name_xyz": ["кабачок", "баклажан", "огурец"],
    "xyz": ["X", "Y", "Z"]
})
```

```python
abc_xyz: DataFrame = abc.merge(xyz, left_on="name", right_on="name_xyz", how="outer")
abc_xyz
```

```python
abc_xyz["name"] = np.where(abc_xyz["name"].isnull(), abc_xyz["name_xyz"], abc_xyz["name"])
abc_xyz
```

```python
abc_xyz.drop("name_xyz", axis=1)
```

## Сохранение датафреймов

```python
df: DataFrame = pd.read_csv(
    "data/data/data.csv",
    encoding="1251"
)

df
```

```python
shops: List[int] = df["DR_Apt"].unique().tolist()
shops

for shop in shops:
    small_df = df[df["DR_Apt"] == shop]
    small_df.to_excel(
        f"data/export/{shop}.xlsx", 
        sheet_name=str(shop), 
        index=False
    )
```

## Загрузка датафреймов из файлов, объединение

```python
# Прочитает данные из Excel-файла
df_xls: DataFrame = pd.read_excel('data/export/2.xlsx')
df_xls
```

```python
glob.glob('data/export/*.xlsx')
```

```python
files: List[str] = glob.glob('data/export/*.xlsx')
dataframes: List[DataFrame] = []

for filename in files:
    part_df = pd.read_excel(filename)
    dataframes.append(part_df)

unioned_pd = pd.concat(dataframes, ignore_index=True)
```

```python
for i, row in unioned_pd.iterrows():
    print("Номер строки:", i, "Дата:", row['DR_Dat'])
    if i == 10:
        break
```
