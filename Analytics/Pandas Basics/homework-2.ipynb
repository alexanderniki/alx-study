## Задание

Загрузите файл в датафрейм Pandas (с именем df).

1. Создайте столбец weekday, в котором будет храниться день недели в русском варианте (Понедельник, Вторник, ...)

2. Создайте столбец date, в котором будет храниться дата регистрации пользователя (без времени)

3. Создайте датафрейм res, который получится в результате группировки исходного датафрейма по дням недели и расчета:

   - Количества регистраций в этот день недели
   - Количества уникальных дат, в которые были регистрации для этого дня недели

   Важно: Столбец с днями недели должен быть не в индексе, а просто столбцом датафрейма.

```python
df: pd.DataFrame = pd.read_csv(
    "data/data/itresume-users-pandas.csv", 
    delimiter=";",
    header=1
)

df.head()
```

```python
# Создаем новой поле с локализованным днем недели

weekday_map: Dict[str, str] = {
    "Monday": "Понедельник",
    "Tuesday": "Вторник",
    "Wednesday": "Среда",
    "Thursday": "Четверг",
    "Friday": "Пятница",
    "Saturday": "Суббота",
    "Sunday": "Воскресенье",
}

df["weekday"] = pd.to_datetime(df["date_joined"]).dt.day_name().map(weekday_map)

df.head()
```

```python
# Создаем новое поле с датой
df["date"] = pd.to_datetime(df["date_joined"]).dt.strftime("%Y-%m-%d")

df.head()
```

```python
res_one: pd.DataFrame = df.groupby(["weekday", "date"]).count()
res_one
```

```python
res: pd.DataFrame = df.groupby(["weekday"]).agg({
    "date": "nunique",
    "id": "count"
}).reset_index()
res
```
