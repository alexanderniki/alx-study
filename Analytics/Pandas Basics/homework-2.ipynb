## Задание 1

Загрузите файл в датафрейм Pandas (с именем df).

1. Создайте столбец weekday, в котором будет храниться день недели в русском варианте (Понедельник, Вторник, ...)

2. Создайте столбец date, в котором будет храниться дата регистрации пользователя (без времени)

3. Создайте датафрейм res, который получится в результате группировки исходного датафрейма по дням недели и расчета:

   - Количества регистраций в этот день недели
   - Количества уникальных дат, в которые были регистрации для этого дня недели

   Важно: Столбец с днями недели должен быть не в индексе, а просто столбцом датафрейма.

```python
df: pd.DataFrame = pd.read_csv(
    "data/data/itresume-users-pandas.csv", 
    delimiter=";",
    header=1
)

df.head()
```

```python
# Создаем новой поле с локализованным днем недели

weekday_map = {
    "Monday": "Понедельник",
    "Tuesday": "Вторник",
    "Wednesday": "Среда",
    "Thursday": "Четверг",
    "Friday": "Пятница",
    "Saturday": "Суббота",
    "Sunday": "Воскресенье",
}

df["weekday"] = pd.to_datetime(df["date_joined"]).dt.day_name().map(weekday_map)

df.head()
```

```python
# Создаем новое поле с датой
df["date"] = pd.to_datetime(df["date_joined"]).dt.date

df.head()
```

```python
res_one: pd.DataFrame = df.groupby(["weekday", "date"]).count()
res_one
```

```python
res = df.groupby("weekday").agg({
    "date": ["count", "nunique"]
}).reset_index()
res
```

## Задание 2

1. Отфильтруйте только тех пользователей, у которых id >= 94. Пользователи с меньшим id - фейковые, они сбивают статистику.
2. Также уберите пользователей с одним из этих ников: 'fnoxonsnbkbxsnb', 'Jonns21241412', 'fbxbnsn'. Это также фейковые юзеры.
3. Создайте столбец cohort, который будет содержать год и месяц регистрации в формате YYYY-MM. 
4. Создайте переменную res, которая будет содержать долю каждой когорты в общем количестве зарегистрированных пользователей. В сумме все доли должны давать 100.

```python
df: pd.DataFrame = pd.read_csv(
    "data/data/itresume-users-pandas.csv", 
    delimiter=";",
    header=1
)

df.head()
```

```python
# Отфильтруем пользователей по id и никам

#df = df[df["id"]>=94 & ~df.loc[:, 'username'].isin(["fnoxonsnbkbxsnb", "Jonns21241412", "fbxbnsn"])]


df = df[df["id"]>=94]
df = df[~df.loc[:, "username"].isin(["fnoxonsnbkbxsnb", "Jonns21241412", "fbxbnsn"])]
df
```

```python
# Создаем поле с датами

df["cohort"] = pd.to_datetime(df["date_joined"]).dt.strftime("%Y-%m")
df
```

```python
# Расчитаем долю каждой когорты в общем количестве 
# зарегистрированных пользователей.

amount: int = df["id"].count()  # Общее количество пользователей

res = df[["cohort", "id"]].groupby("cohort").count()
res = (res["id"] / amount * 100)
res.name = "proportion"
```

```python
print(df)
print(df.index.to_list())
print(res)
```

## Задание 3


1. Создайте столбец cohort, который будет содержать год и месяц регистрации в формате YYYY-MM.
2. Создайте столбец language, который проводит соответствие между языком и language_id по правилу: Python -> 3, SQL -> 2
3. Посчитаем - в скольких задачах в среднем один пользователь отправлял на выполнение код в каждой когорте по каждому языку.

```python
coderun: pd.DataFrame = pd.read_csv(
    "data/data/itresume-coderun.csv", 
    delimiter=",",
    header=0
)

coderun.head(10)
```

```python
# Создаем новое поле
coderun["cohort"] = pd.to_datetime(coderun["created_at"]).dt.strftime("%Y-%m")
coderun
```

```python
# Смапим коды языков на их название

langs = {
    3: "Python",
    2: "SQL"
}

coderun["language"] = coderun["language_id"].map(langs)
coderun
```

```python
# Уберем данные до 2022-01

coderun_filtered: pd.DataFrame = coderun.copy()
coderun_filtered["created_at"] = pd.to_datetime(coderun_filtered["created_at"])
coderun_filtered = coderun_filtered[coderun_filtered["created_at"] >= "2022-01-01"]
```

```python
tasks_per_user: pd.DataFrame = (
    coderun_filtered.groupby(["cohort", "language", "user_id"])["problem_id"]
    .nunique()
    .reset_index()
)

print(tasks_per_user)

df: pd.DataFrame = tasks_per_user.pivot_table(
    index="cohort",
    columns="language",
    values="problem_id",
    #aggfunc="mean"
)

print(df)
```
