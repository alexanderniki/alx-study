### Как задавать фильтры

```python
# через метод query

sales.query('category=="Аптека"')  # оставить строки с категорией Аптека
```

```python
# индексированный фильтр

data_amur = sales[sales['region'] == 'Амурская область']
data_amur.head()
```

```python
# если надо сразу несколько регионов отфильтровать

data_east = sales[sales['region'].isin(['Амурская область', 'Хабаровский край', 'Приморский край',
                                        'Иркутская область', 'Красноярский край'])]

data_east.head()
```

```python
# несколько фильтров типа И (одновременно выполняются ВСЕ условия)
sales[(sales['region']=='Хабаровский край') & (sales['discount'] > 0.05)]
```

### Распределение продаж по категориям
Для каждой категории посчитаем сумму продаж по столбцу sales

```python
sales.groupby('category').agg({'sales': 'sum'})  # для каждой категории считаем сумму столбца sales
```

### Самые активные покупатели
Действуем аналогично, только группировка по ID покупателя + добавим сортировку результата по убыванию.

```python
sales.groupby('customer_id').sum()  # посчитаются ВСЕ столбцы
```

```python
sales.groupby('customer_id').agg({'sales': 'sum'}).sort_values('sales', ascending=False).head()
```

```python
# как получить топ-10 самых активных покупателей

top_customers = sales.groupby('customer_id').agg({'sales': 'sum'}).sort_values('sales', ascending=False).head().index
top_customers
```

```python
# записать результат в Эксель
top_customers.to_excel('result.xlsx')
```

## Таблица возвратов

```python
returns = pd.read_csv('/content/drive/My Drive/Colab Notebooks/returns.csv')
returns.head()
```

```python
returns.info()
```

Объединяем таблицы продаж и возвратов

```python
%%time

joined = sales.merge(returns, on='order_id')  # how - тип объединения (left, right)
joined.head()
```

Построим сводную таблицу, где для каждой категории посмотрим количество строк для возвратов (1) и успешных заказов (0).

```python
pivot = joined.pivot_table(index='category', columns='return', values='sales', aggfunc='count')
pivot
```

Считаем долю возвратов

```python
pivot['returns_share'] = pivot[1] / (pivot[0] + pivot[1])
pivot
```
