2. Определение регионов с наибольшими продажами и наименьшими возвратами в категориях «Бытовая техника» и «Одежда и обувь»

- Объедините таблицы из файлов sales.csv и returns.csv по общему столбцу order_id.
- Отфильтруйте полученную таблицу по категориям «Бытовая техника» и «Одежда и обувь».
- Посчитайте, в каком регионе были наибольшие продажи.
- Посчитайте долю возвратов товаров для каждого региона.

```python
# Объединяем таблицы продаж и возвратов
sales_full = sales.merge(returns, on='order_id')
sales_full.head()

# Фильтруем по категориям
data_cat =  sales_full[sales_full['category'].isin(['Бытовая техника', 'Одежда и обувь'])]
data_cat.head(10)

# Считаем продажи по регионам и ищем максимум
data_reg_sales = data_cat.groupby('region').agg({'sales': 'sum'}).sort_values('sales', ascending=False)
data_reg_sales
```

```python
# Считаем количество возвратов по регионам

data_reg_returns = data_cat.groupby('region').agg({'return': 'sum'}).sort_values('return', ascending=False)
data_reg_returns.head()
```

```python
pivot = data_cat.pivot_table(index='region', columns='return', values='sales', aggfunc='count')
pivot['returns_share'] = pivot[1] / (pivot[0] + pivot[1])
pivot
```
